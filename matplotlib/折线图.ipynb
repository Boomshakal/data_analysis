```python
import pandas as pd
unrate = pd.read_csv('unrate1.csv')
unrate['DATE']=pd.to_datetime(unrate['DATE'])
print(unrate.head())
```

```python
import matplotlib.pyplot as plt
plt.plot()
plt.show()
```

```python
first_twelve=unrate[0:12]
plt.plot(first_twelve['DATE'],first_twelve['UNEMPLOYMENTRATE'])
plt.xticks(rotation=45)  #斜着45°
plt.xlabel('month')
plt.ylabel('RATE')
plt.title('UNEMPLOYMENT RATE')
plt.show()
```
