```python
import pandas as pd
import numpy as np
titanic_survival=pd.read_csv('titanic_train.csv')
titanic_survival.head()
```

```python
age=titanic_survival['Age']
print(type(age))
print(age[2:10])
age_is_null=pd.isnull(age)
print(age_is_null)
age_null_true=age[age_is_null]
print(age_null_true)
print(len(age_null_true))
```

```python
mean_age=sum(titanic_survival['Age'])/len(titanic_survival['Age'])
print(mean_age)
```

```python
good_ages=titanic_survival['Age'][age_is_null==False]
correct_age=sum(good_ages)/len(good_ages)
print(correct_age)
```

```python
correct_mean_age=titanic_survival['Age'].mean()
print(correct_mean_age)
```

```python
pclasses=[1,2,3]
fare_by_class={}
for pclass in pclasses:
    pclass_rows=titanic_survival[titanic_survival['Pclass']==pclass]
    pclass_fares=pclass_rows['Fare']
    fare_for_class=pclass_fares.mean()
    fare_by_class[pclass]=fare_for_class
print(fare_by_class)
```

```python
#数据透视表
passenger_survial=titanic_survival.pivot_table(index='Pclass',values='Survived',aggfunc=np.mean)
print(passenger_survial)
```

```python
passenger_age=titanic_survival.pivot_table(index='Pclass',values='Age')  #默认求均值
print(passenger_age) 
```

```python
port_stats=titanic_survival.pivot_table(index='Embarked',values=['Fare','Survived'],aggfunc=np.sum)
print(port_stats)
```

```python
drop_na_columns=titanic_survival.dropna(axis=1)  #空的列去掉
print(drop_na_columns)
new_titanic_survival=titanic_survival.dropna(axis=0,subset=['Age','Sex'])  #空的行去掉
print(new_titanic_survival)
```

```python
row_index_83_age=titanic_survival.loc[83,'Age']
row_index_228_Pclass=titanic_survival.loc[228,'Pclass']
print(row_index_228_Pclass,row_index_83_age)
```

```python
new_titanic_survival=titanic_survival.sort_values('Age',ascending=False)
print(new_titanic_survival)
titanic_reindexed=new_titanic_survival.reset_index(drop=True)
print('------------')
print(titanic_reindexed[0:10])
```

```python
def hundredth_row(column):
    hundredth_item=column.loc[99]
    return hundredth_item

hundredth_row=titanic_survival.apply(hundredth_row)
print(hundredth_row)
```

```python
def not_null_count(column):
    column_null=pd.isnull(column)
    null=column[column_null]
    return len(null)
column_null_count=titanic_survival.apply(not_null_count)
print(column_null_count)
```

```python
def which_class(row):
    pclass=row['Pclass']
    if pd.isnull(pclass):
        return 'unknown'
    if pclass==1:
        return 'First class'
    if pclass==2:
        return 'Second class'
    if pclass==3:
        return 'Third class'
classes=titanic_survival.apply(which_class,axis=1)
print(classes)
```

```python
def is_minor(row):
    age=row['Age']
    if age<18:
        return True
    else:
        return False

minors=titanic_survival.apply(is_minor,axis=1)
print(minors)

def generate_age_lable(row):
    age=row['Age']
    if pd.isnull(age):
        return 'unknown'
    elif age<18:
        return 'minor'
    else:
        return 'adult'
age_lables=titanic_survival.apply(generate_age_lable,axis=1)
print(age_lables)
```

```python
titanic_survival['age_lables']=age_lables
ages_survial=titanic_survival.pivot_table(index='age_lables',values='Survived',aggfunc=np.mean)
print(ages_survial)
```
